```python
import astx
```

```python
comp_1 = astx.Comprehension(
    target=astx.Variable('x'),
    iterable=astx.Variable('my_List'),
)
comp_2 = astx.Comprehension(
    target=astx.Variable('x'),
    iterable=astx.Variable('myList'),
    conditions=[
        astx.LiteralBoolean(True)
    ]
)
```

```python
comp_2
```

```python
generators=[
        comp_1, comp_2
    ]
```

```python
generators
```

```python
gen_expr = astx.GeneratorExpr(
    element=astx.BinaryOp(
        op_code="+", lhs=astx.Variable("x"), rhs=astx.Variable("x")
    ),
    generators=[
        comp_1, comp_2
    ]
    
)
```
